```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Функция для варианта 9
def piecewise_function(t):
    result = np.zeros_like(t)
    
    # t <= 0
    mask1 = t <= 0
    result[mask1] = -2 * np.sin(t[mask1])
    
    # 0 < t <= 1
    mask2 = (t > 0) & (t <= 1)
    result[mask2] = t[mask2]
    
    # 1 < t <= 2
    mask3 = (t > 1) & (t <= 2)
    result[mask3] = 1
    
    # 2 < t <= 3
    mask4 = (t > 2) & (t <= 3)
    result[mask4] = 3 - t[mask4]
    
    # t > 3
    mask5 = t > 3
    result[mask5] = 2 * np.sin(t[mask5] - 3)
    
    return result

# Создаем массив значений t
t = np.linspace(-2, 5, 1000)
f = piecewise_function(t)

# Построение графика с разными цветами для каждого участка
plt.figure(figsize=(10, 6))

# Разбиваем на участки для разных цветов
mask1 = t <= 0
mask2 = (t > 0) & (t <= 1)
mask3 = (t > 1) & (t <= 2)
mask4 = (t > 2) & (t <= 3)
mask5 = t > 3

# Рисуем каждый участок своим цветом
plt.plot(t[mask1], f[mask1], 'b-', linewidth=2, label='t ≤ 0: -2sin(t)')
plt.plot(t[mask2], f[mask2], 'r-', linewidth=2, label='0 < t ≤ 1: t')
plt.plot(t[mask3], f[mask3], 'g-', linewidth=2, label='1 < t ≤ 2: 1')
plt.plot(t[mask4], f[mask4], 'm-', linewidth=2, label='2 < t ≤ 3: 3-t')
plt.plot(t[mask5], f[mask5], 'orange', linewidth=2, label='t > 3: 2sin(t-3)')

# Оформление графика
plt.xlabel('t', fontsize=12)
plt.ylabel('f(t)', fontsize=12)
plt.title('График кусочно-заданной функции (Вариант 9)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ЗАДАНИЕ 2 для варианта 9

def voltage_function_v9(t):
    """Функция напряжения для варианта 9: 7⋅sin⁵(t)⋅log(17⋅t)"""
    return 7 * (np.sin(t))**5 * np.log(17 * t)

# Диапазон значений t: 0...2π
t2 = np.linspace(0.01, 2*np.pi, 1000)  # начинаем с 0.01 чтобы избежать log(0)

# Вычисляем значения напряжения
U = voltage_function_v9(t2)

# Заданные точки по шкале t
points_t = [0.5, 1, 1.5]
points_U = voltage_function_v9(np.array(points_t))

# Находим максимальное значение напряжения
max_U = np.max(U)
max_t = t2[np.argmax(U)]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(t2, U, 'r-', linewidth=2, label='U(t) = 7⋅sin⁵(t)⋅log(17⋅t)')

# Отмечаем заданные точки
plt.scatter(points_t, points_U, color='green', s=80, zorder=5, label='Заданные точки')

# Отмечаем максимальное значение
plt.scatter(max_t, max_U, color='blue', s=100, zorder=5, label=f'Максимум: U({max_t:.2f}) = {max_U:.2f}')

plt.xlabel('Время t', fontsize=12)
plt.ylabel('Напряжение U(t)', fontsize=12)
plt.title('График напряжения от времени (Вариант 9)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()

# Выводим значения в заданных точках
print("Значения напряжения в заданных точках:")
for i, (t_val, U_val) in enumerate(zip(points_t, points_U)):
    print(f"t = {t_val}: U({t_val}) = {U_val:.4f}")

print(f"\nМаксимальное значение напряжения: U({max_t:.4f}) = {max_U:.4f}")

plt.show()
```
