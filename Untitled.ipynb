```python
#1
import matplotlib.pyplot as plt
import numpy as np


# Ввод параметров
a = float(input("Введите длину катета a: "))
c = float(input("Введите длину гипотенузы c: "))


# Проверка корректности
if c <= a:
    print("Гипотенуза должна быть больше катета.")
else:
    # Вычисление второго катета b с помощью теоремы Пифагора
    b = (c**2 - a**2)**0.5


    # Координаты вершин треугольника
    triangle_points = np.array([[0, 0], [a, 0], [0, b], [0, 0]])


    # Построение треугольника
    plt.fill(triangle_points[:, 0], triangle_points[:, 1])


    # Настройки графика
    plt.xlim(-1, a + 1)
    plt.ylim(-1, b + 1)
    plt.title('Прямоугольный треугольник')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid()


    plt.show()
```

```python
#2
import matplotlib.pyplot as plt
import math


# Ввод координат центра и длины стороны треугольника
x = float(input("Введите координату x центра треугольника: "))
y = float(input("Введите координату y центра треугольника: "))
a = float(input("Введите длину стороны треугольника: "))


# Вычисление высоты и координат вершин треугольника
h = (math.sqrt(3) / 2) * a  # Высота равностороннего треугольника
x1 = x - a / 2               # Первая вершина (лево-нижняя)
y1 = y - h / 3               # Первая вершина (лево-нижняя)
x2 = x + a / 2               # Вторая вершина (право-нижняя)
y2 = y - h / 3               # Вторая вершина (право-нижняя)
x3 = x                       # Третья вершина (верхняя)
y3 = y + (2 * h) / 3        # Третья вершина (верхняя)


# Настройка графика
plt.figure()
plt.grid()
plt.title("Равносторонний треугольник")
plt.xlabel("x")
plt.ylabel("y")


# Построение треугольника
plt.plot([x1, x2, x3, x1], [y1, y2, y3, y1], marker='o')  # Добавлен маркер для вершин треугольника
plt.scatter(x, y, color='red', label='Центр треугольника')  # Отображение центра треугольника


# Добавление легенды
plt.legend()


# Показ графика
plt.axis('equal')  # Установка равных масштабов по осям для правильного отображения треугольника
plt.show()
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt


# Ввод значения N и x
N = int(input("Введите количество членов ряда N: "))
x = float(input("Введите значение x: "))


# Расчет суммы первых N членов ряда
S = sum((-1)**n * x**n for n in range(N))


# Вывод значения S
print(f"Сумма первых {N} членов ряда для x = {x}: S = {S}")


# Определение функции y(x)
def y(x):
    return 1 / (1 + x)


# Создание массива значений x для построения графика
x_values = np.linspace(-10, 10, 400)
y_values = y(x_values)


# Построение графика
plt.plot(x_values, y_values, label='y(x) = 1/(1 + x)', color='blue')
plt.scatter(x, S, color='red', label=f'Point (x={x}, S={S})')
plt.title('График функции y(x) и точка (x, S)')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black', lw=0.5, ls='--')
plt.axvline(0, color='black', lw=0.5, ls='--')
plt.legend()
plt.grid()
plt.show()
```

```python
#9
import numpy as np
import math
import matplotlib.pyplot as plt
 
x = int(input('Введите X: '))
S = 1
n = int(input('Введите число: '))
for i in range(1, n):
    d = (x ** i) / math.factorial(i)
    S += d
t = np.arange(0, 6, 0.1)
y = np.exp(t)
 
plt.plot(t, y, label='y = exp(t)')
plt.plot(x, S, 'ro', label='Сумма ряда')
plt.title('График экспоненты и сумма ряда')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```

```python
#12
import matplotlib.pyplot as plt
def calculate_sum(N):
  S_values = []
  sum = 0
  for k in range(1, N + 1):
    sum += k** 2
    S_values.append(sum) # Добавляем текущее значение суммы в список
  return S_values
def main():
  N = int(input("Введите N: "))
  S_values = calculate_sum(N)
  plt.plot(range (1, N + 1), S_values, marker='o')
  plt.xlabel('k')
  plt.ylabel('s')
  plt.title('График зависимости S(k)')
  plt.show()
if __name__ == "__main__":
  main()
```

```python
#13
import numpy as np
import matplotlib.pyplot as plt


# Генерируем массив m
m = np.random.randint(1, 100, size=20)  # Массив из 20 случайных чисел от 1 до 100
print("Массив m:", m)


# Находим элементы, кратные трём
elements = m[m % 3 == 0]


# Рассчитываем среднее арифметическое
if len(elements) > 0:
    average = np.mean(elements)
    print("Среднее арифметическое элементов, кратных трём:", average)
else:
    average = None
    print("Нет элементов, кратных трём.")


# Строим график
plt.figure(figsize=(10, 6))
plt.plot(m, marker='o', label='Элементы массива m')


# Выделяем элементы, кратные трём
for i, value in enumerate(m):
    if value % 3 == 0:
        plt.scatter(i, value, color='red', s=100, label='Кратные трём' if i == 0 else "")
       
plt.title('График элементов массива m')
plt.xlabel('Номер элемента массива (k)')
plt.ylabel('Значение элемента M(k)')
plt.legend()
plt.grid()
plt.show()


```
