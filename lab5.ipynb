```python
for i in range(1, 5):
    # Открытие файла для чтения с указанием кодировки
    filename = f'{i}.txt'
    f = open(filename, 'r', encoding='utf-8')
    
    # Чтение данных...
    G1 = int(f.readline())
    G2 = int(f.readline())
    S1 = int(f.readline())
    S2 = int(f.readline())
    S3 = int(f.readline())
    S4 = int(f.readline())
    VL1 = int(f.readline())
    VL2 = int(f.readline())
    KL1 = int(f.readline())
    KL2 = int(f.readline())
    KL3 = int(f.readline())
    
    f.close()
    
    path1 = G1 and S1 and KL1 and VL2 and S3 and KL3
    path2 = G2 and S4 and KL2 and VL2 and S3 and KL3
    
    if path1 or path2:
        result = "Есть электроснабжение у П2"
    else:
        result = "Нет электроснабжения у П2"
    
    # Запись с указанием кодировки UTF-8
    result_filename = f'result{i}.txt'
    f = open(result_filename, 'w', encoding='utf-8')
    f.write(result)
    f.close()
    
    print(f"Файл {i}.txt: {result}")
```
