```python
import matplotlib.pyplot as plt
import numpy as np

# Координаты точки A (Вариант 11)
# x = 25, y = 30, z = 10
x, y, z = 25, 30, 10

# --- Часть 1: 3D Визуализация ---
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Основная точка A
ax.scatter(x, y, z, color='red', s=100, label=f'A({x}, {y}, {z})')
ax.text(x, y, z+3, f'A({x}, {y}, {z})', color='red', fontsize=12, fontweight='bold')

# Горизонтальная проекция A1 (на плоскость XY)
ax.scatter(x, y, 0, color='blue', s=80, marker='x', label=f'A₁({x}, {y}, 0)')
ax.text(x, y, -3, f'A₁({x}, {y}, 0)', color='blue', fontsize=10)

# Фронтальная проекция A2 (на плоскость XZ)
ax.scatter(x, 0, z, color='green', s=80, marker='x', label=f'A₂({x}, 0, {z})')
ax.text(x, -3, z, f'A₂({x}, 0, {z})', color='green', fontsize=10)

# Профильная проекция A3 (на плоскость YZ)
ax.scatter(0, y, z, color='purple', s=80, marker='x', label=f'A₃(0, {y}, {z})')
ax.text(-3, y, z, f'A₃(0, {y}, {z})', color='purple', fontsize=10)

# Линии проекционной связи
ax.plot([x, x], [y, y], [z, 0], 'k--', linewidth=0.8)  # A-A1
ax.plot([x, x], [y, 0], [z, z], 'k--', linewidth=0.8)  # A-A2
ax.plot([x, 0], [y, y], [z, z], 'k--', linewidth=0.8)  # A-A3

# Оси координат
ax.axhline(y=0, color='black', linewidth=1)
ax.plot([0, 0], [0, 0], [0, 50], 'black', linewidth=1)

# Подписи осей
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.set_title(f'ЗАДАНИЕ (Вариант 11): Точка A({x}, {y}, {z}) и её проекции', fontsize=14, fontweight='bold')

# Границы осей
ax.set_xlim(0, 50)
ax.set_ylim(0, 40)
ax.set_zlim(0, 40)

plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# --- Часть 2: Эпюр (комплексный чертеж) ---
fig2, axs = plt.subplots(2, 2, figsize=(12, 10))
fig2.suptitle(f'ЭПЮР ТОЧКИ (Вариант 11) A({x}, {y}, {z})', fontsize=16, fontweight='bold')

# Горизонтальная проекция (XY)
axs[0, 0].scatter(x, y, color='blue', s=100)
axs[0, 0].text(x+2, y+2, f'A₁({x}, {y})', color='blue', fontsize=12)
axs[0, 0].axhline(y=0, color='black', linewidth=0.8)
axs[0, 0].axvline(x=0, color='black', linewidth=0.8)
axs[0, 0].set_title('Горизонтальная проекция (XY)', fontsize=11)
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')
axs[0, 0].grid(True)
axs[0, 0].set_xlim(0, 50)
axs[0, 0].set_ylim(0, 40)

# Фронтальная проекция (XZ)
axs[0, 1].scatter(x, z, color='green', s=100)
axs[0, 1].text(x+2, z+2, f'A₂({x}, {z})', color='green', fontsize=12)
axs[0, 1].axhline(y=0, color='black', linewidth=0.8)
axs[0, 1].axvline(x=0, color='black', linewidth=0.8)
axs[0, 1].set_title('Фронтальная проекция (XZ)', fontsize=11)
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Z')
axs[0, 1].grid(True)
axs[0, 1].set_xlim(0, 50)
axs[0, 1].set_ylim(0, 40)

# Профильная проекция (YZ)
axs[1, 0].scatter(y, z, color='purple', s=100)
axs[1, 0].text(y+2, z+2, f'A₃({y}, {z})', color='purple', fontsize=12)
axs[1, 0].axhline(y=0, color='black', linewidth=0.8)
axs[1, 0].axvline(x=0, color='black', linewidth=0.8)
axs[1, 0].set_title('Профильная проекция (YZ)', fontsize=11)
axs[1, 0].set_xlabel('Y')
axs[1, 0].set_ylabel('Z')
axs[1, 0].grid(True)
axs[1, 0].set_xlim(0, 40)
axs[1, 0].set_ylim(0, 40)

# Комплексный чертеж (эпюр)
axs[1, 1].axis('off')
axs[1, 1].set_xlim(0, 100)
axs[1, 1].set_ylim(0, 100)

# Оси эпюра (смещение, чтобы точки влезли)
axs[1, 1].axhline(y=50, color='black', linewidth=1.5)  # Ось X
axs[1, 1].axvline(x=50, color='black', linewidth=1.5)  # Ось Z

# Точки на эпюре
axs[1, 1].scatter(50+x, 50-y, color='blue', s=100, label=f'A₁')
axs[1, 1].scatter(50+x, 50+z, color='green', s=100, label=f'A₂')
axs[1, 1].scatter(50-y, 50+z, color='purple', s=100, label=f'A₃')

# Подписи
axs[1, 1].text(50+x+3, 50-y+3, f'A₁({x},{y})', color='blue', fontsize=9)
axs[1, 1].text(50+x+3, 50+z+3, f'A₂({x},{z})', color='green', fontsize=9)
axs[1, 1].text(50-y-15, 50+z+3, f'A₃({y},{z})', color='purple', fontsize=9)

# Линии связи
axs[1, 1].plot([50+x, 50+x], [50-y, 50+z], 'k--', linewidth=0.8)
axs[1, 1].plot([50+x, 50-y], [50+z, 50+z], 'k--', linewidth=0.8)

axs[1, 1].text(50, 95, 'Z', fontsize=12, fontweight='bold')
axs[1, 1].text(95, 50, 'X', fontsize=12, fontweight='bold')
axs[1, 1].set_title('Комплексный чертеж', fontsize=11)
axs[1, 1].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
