```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(-2, 6, 1000)
f = np.zeros_like(t)

# Создаем маски для каждого участка функции
mask1 = t <= 0
mask2 = (t > 0) & (t <= 1)
mask3 = (t > 1) & (t <= 2)
mask4 = (t > 2) & (t <= 3)
mask5 = t > 3

# Вычисляем значения для каждого участка
f1 = -2 * np.sin(t[mask1])
f2 = t[mask2]
f3 = np.ones_like(t[mask3])
f4 = 3 - t[mask4]
f5 = 2 * np.sin(t[mask5] - 3)

# Создаем дополнительные функции для сравнения
g = np.sin(t)  # Синусоида
h = np.exp(-0.5 * t) * 2  # Затухающая экспонента

plt.figure(figsize=(12, 8))

# Рисуем кусочную функцию разноцветными сегментами
plt.plot(t[mask1], f1, 'b-', linewidth=3, label='f(t) при t ≤ 0', alpha=0.8)
plt.plot(t[mask2], f2, 'r-', linewidth=3, label='f(t) при 0 < t ≤ 1', alpha=0.8)
plt.plot(t[mask3], f3, 'g-', linewidth=3, label='f(t) при 1 < t ≤ 2', alpha=0.8)
plt.plot(t[mask4], f4, 'purple', linewidth=3, label='f(t) при 2 < t ≤ 3', alpha=0.8)
plt.plot(t[mask5], f5, 'orange', linewidth=3, label='f(t) при t > 3', alpha=0.8)

# Добавляем другие функции для сравнения
plt.plot(t, g, 'k--', linewidth=2, label='g(t) = sin(t)')
plt.plot(t, h, 'm:', linewidth=2, label='h(t) = 2e^(-t/2)')

# Добавляем вертикальные линии в точках разрыва
plt.axvline(x=0, color='gray', linestyle=':', alpha=0.5)
plt.axvline(x=1, color='gray', linestyle=':', alpha=0.5)
plt.axvline(x=2, color='gray', linestyle=':', alpha=0.5)
plt.axvline(x=3, color='gray', linestyle=':', alpha=0.5)

plt.xlabel('t', fontsize=12)
plt.ylabel('f(t)', fontsize=12)
plt.title('Разноцветный график кусочной функции и других функций', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Сохраняем график
plt.savefig('images/plot_9_1.png', dpi=300, bbox_inches='tight')
plt.show()
```

```python
t = np.linspace(0.01, 2 * np.pi, 1000)
U = 7 * (np.sin(t))**5 * np.log(17 * t)

max_U = np.max(U)
t_max = t[np.argmax(U)]

t_points = [0.5, 1, 1.5]
U_points = 7 * (np.sin(t_points))**5 * np.log(17 * np.array(t_points))

plt.figure(figsize=(10, 6))
plt.plot(t, U, linewidth=2, label='U(t)')
plt.scatter(t_points, U_points, color='red', zorder=5, label='Заданные точки')
plt.scatter(t_max, max_U, color='green', marker='*', s=200, label=f'Максимум: t={t_max:.2f}, U={max_U:.2f}')
plt.xlabel('Время t')
plt.ylabel('Напряжение U(t)')
plt.grid(True)
plt.legend()
plt.savefig('images/plot_9_2.png')
```

```python
# Вывод значений в точках
for tp, Up in zip(t_points, U_points):
    print(f't = {tp}: U(t) = {Up:.4f}')
```
