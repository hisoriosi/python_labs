```python
import numpy as np
import random

# Данные для варианта 9
allowed_strikes = 10  # Допустимое число ударов
num_simulations = 1000

# Координаты подстанции (твоя Г-образная форма)
x = [50, 50, 100, 100, 250, 250, 350, 350, 50]
y = [50, 150, 150, 200, 200, 150, 150, 50, 50]

# Простой поиск максимальной плотности
max_density = 0

for test_density in range(1, 500):  # Проверяем плотности от 1 до 500
    total_hits = 0
    
    # 1000 повторений моделирования
    for simulation in range(num_simulations):
        hits = 0
        # Моделируем test_density ударов молнии
        for strike in range(test_density):
            # Случайные координаты удара
            x_strike = random.uniform(0, 1000)
            y_strike = random.uniform(0, 1000)
            
            # Простая проверка попадания (можно улучшить для Г-образной формы)
            if (50 <= x_strike <= 350) and (50 <= y_strike <= 200):
                # Исключаем правый верхний угол (сложную часть можно упростить)
                if not (100 < x_strike < 250 and 150 < y_strike < 200):
                    hits += 1
        
        total_hits += hits
    
    # Среднее количество ударов за все симуляции
    average_hits = total_hits / num_simulations
    
    # Проверяем условие
    if average_hits <= allowed_strikes:
        max_density = test_density
    else:
        break

print(f"Максимально допустимая плотность: {max_density} ударов/км²/год")
print(f"При плотности {max_density} среднее количество ударов: {average_hits:.2f}")
print(f"Допустимое количество ударов: {allowed_strikes}")
```
