```python
import numpy as np
```

```python
N = 5

# 2 четверть
A_1 = np.zeros((N,N))
A_1[0:5, 0:5] = np.arange(10, 5, -1)

# 3 четверть
H = 10
A_2 = np.zeros((N,N))
for k in range(N):
  A_2[:N-k,k] = H
  H -= 1

# 4 четверть
A_3 = np.zeros((N,N))

# 1 четверть
N = 5
A_4 = np.zeros((N,N))
for k in range(N):
  A_4[k:, k] = np.arange(k+1,N+1)

# объединение 1 и 2 четверти и 3 и 4 
up = np.column_stack((A_1, A_4))
down = np.column_stack((A_2, A_3))

# полное объединение матрицы
zero_matrix = np.zeros((10, 10), dtype=int)
zero_matrix[0:5, 0:10] = up
zero_matrix[5:10, 0:10] = down
zero_matrix
```
