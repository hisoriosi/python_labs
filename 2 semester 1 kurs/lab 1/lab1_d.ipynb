```python
import numpy as np
import time
```

```python
def f(x):
    return x**3 + 2.5*x**2 - 6.5*x - 15

def phi(x):
    return (x**3 + 2.5*x**2 - 15) / 6.5
```

```python
def simple_iteration_method(x, eps_value: float = 0.0001):  
    iterations = 0
    x_prev = x
    while True:
        x_next = phi(x_prev)
        if abs(x_next - x_prev) < eps_value: 
            print(f'Итераций: {iterations}')
            return x_next
        x_prev = x_next
        iterations += 1
    return x_next
```

```python
def get_boundaries(start: float, finish: float):
    full_range = list(np.arange(start, finish, 0.5))
    x_values = []
    roots = []
    for v in full_range:
        try:
            result = simple_iteration_method(v)
            x_values.append(v.item())
            roots.append(result.item())
        except BaseException:
            continue
    print(x_values)    
    print(roots)
```
