```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

# Исходные данные
H = [0, 100, 2500]
B = [0, 1.6, 1.8]

# Интерполяционная функция
f = interp1d(H, B, kind='linear')

# Значение при 1000 А/м
H_target = 1000
B_target = f(H_target)
print(f"B при H = 1000 А/м: {B_target:.4f} Тл")

# Массивы для графика
H_smooth = np.linspace(0, 2500, 500)
B_smooth = f(H_smooth)

# График
plt.figure(figsize=(8, 5))
plt.plot(H_smooth, B_smooth, 'b-')
plt.plot(H, B, 'ro')
plt.plot(H_target, B_target, 'gs', markersize=8)
plt.grid(True)
plt.show()
```
