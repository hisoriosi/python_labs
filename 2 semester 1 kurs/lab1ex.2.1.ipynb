```python
import matplotlib.pyplot as plt
import numpy as np
# Объявление функции, содержащей правую часть уравнения
def my_fun(I):
  E = 10
  r = 2
  return I - E/(r+10/(1+I**2))
# Создание массивов координат точек графика
I = np.arange(1,5,0.01)
Y = my_fun(I)
# Построение графика
plt.figure(1)
plt.plot(I,Y)
plt.grid()
```

```python
from scipy.optimize import bisect
I = bisect(my_fun,3.5,4)
print('Расчёт тока схемы I = {:.2f}'.format(I))
```
